```python
import matplotlib.pyplot as plt
```

```python
class VehicleInsurance:
    def __init__(self, vehicle_value, insurance_years, annual_premium_rate):
        self.initial_value = vehicle_value
        self.insurance_years = insurance_years
        self.annual_premium_rate = annual_premium_rate
        self.depreciation_rate = 0.07  # 7% annual depreciation

    def calculate_premiums(self):
        current_value = self.initial_value
        premiums = []

        for year in range(1, self.insurance_years + 1):
            annual_premium = current_value * self.annual_premium_rate
            monthly_premium = annual_premium / 12
            quarterly_premium = annual_premium / 4

            premiums.append({
                'year': year,
                'vehicle_value': current_value,
                'annual_premium': annual_premium,
                'monthly_premium': monthly_premium,
                'quarterly_premium': quarterly_premium
            })

            current_value *= (1 - self.depreciation_rate)

        return premiums

    def plot_premium_chart(self):
        premiums = self.calculate_premiums()
        years = [p['year'] for p in premiums]
        annual_premiums = [p['annual_premium'] for p in premiums]
        monthly_premiums = [p['monthly_premium'] for p in premiums]
        quarterly_premiums = [p['quarterly_premium'] for p in premiums]

        plt.figure(figsize=(12, 6))
        plt.plot(years, annual_premiums, label='Annual Premium')
        plt.plot(years, [m * 12 for m in monthly_premiums], label='Monthly Premium (Annualized)')
        plt.plot(years, [q * 4 for q in quarterly_premiums], label='Quarterly Premium (Annualized)')
        plt.title('Insurance Premiums Over Time')
        plt.xlabel('Year')
        plt.ylabel('Premium Amount')
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
# Example usage
insurance = VehicleInsurance(vehicle_value=50000, insurance_years=10, annual_premium_rate=0.05)
premiums = insurance.calculate_premiums()
```

```python
for premium in premiums:
    print(f"Year {premium['year']}:")
    print(f"  Vehicle Value: ${premium['vehicle_value']:.2f}")
    print(f"  Annual Premium: ${premium['annual_premium']:.2f}")
    print(f"  Monthly Premium: ${premium['monthly_premium']:.2f}")
    print(f"  Quarterly Premium: ${premium['quarterly_premium']:.2f}")
    print()
```

```python
insurance.plot_premium_chart()
```
